### Generate a t-distribution

```python
from scipy.stats import t
import matplotlib.pyplot as plt
import seaborn as sns

rv_numbers = t.rvs(size = 10, df = 4)
sns.histplot(rv_numbers)
plt.show()
```

As the sample size approaches considerably large value, the t-distribution converges to become Normal.

```python
rv_numbers = t.rvs(size = 100000, df = 4)
sns.histplot(rv_numbers)
plt.show()
```

### Visualizing t--distribution

```python
from scipy.stats import t
from scipy.stats import norm
import numpy as np
import seaborn as sns

# List of possible outcomes
outcomes = np.linspace(-4,4,100)

# List of Probability of each possible outcome (PMF)
pdf_values_norm = [norm.pdf(i, loc = 0, scale = 1) for i in outcomes]
pdf_values_1 = [t.pdf(i, df=1) for i in outcomes]
pdf_values_4 = [t.pdf(i, df=4) for i in outcomes]
pdf_values_20 = [t.pdf(i, df=20) for i in outcomes]


# Plotting the graph
plt.subplots(figsize = (10,5))
sns.lineplot(x = outcomes, y = pdf_values_norm, color = 'purple')
sns.lineplot(x = outcomes, y = pdf_values_1, color = 'red')
sns.lineplot(x = outcomes, y = pdf_values_4, color = 'orange')
sns.lineplot(x = outcomes, y = pdf_values_20, color = 'yellow')
plt.show()
```
