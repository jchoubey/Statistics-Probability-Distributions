### Generate a Normal Distribution

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns

rv_numbers = norm.rvs(loc = 0, scale = 1, size = 10000)
sns.histplot(rv_numbers)
plt.show()
```

This is a standard normal distribution with mean = 0 & standard deviation = 1.

### Probabilities of a Normal Distribution

```python
norm.pdf(x = 0, loc = 0, scale = 1)
```

```python
norm.cdf(x = 0, loc = 0, scale = 1)
```

### Visualizing a Normal Distribution

```python
from scipy.stats import norm
import numpy as np
import seaborn as sns

loc = 0
scale = 1

# List of possible outcomes
outcomes = np.linspace(-4,4,100)

# List of Probability of each possible outcome (PMF)
pdf_values = [norm.pdf(i, loc, scale) for i in outcomes]

# Plotting the graph
plt.subplots(figsize = (10,5))
sns.lineplot(x = outcomes, y = pdf_values)
plt.show()
```

### Example Problems

In almost all Normal Distribution problems, the provided normal distribution is first converted to standard normal distribution by doing the following:
1. Shifting the mean to 0 (subtracting actual mean from all x values)
2. Scaling the standard deviation to 1 (dividing by actual standard deviation)

After doing tehse 2 steps, each x positional variable becomes a z-score: a variable on the x-axis of the standardized normal distribution. It then becomes easy to calculate the area under the curve at that point (z-score) which is inturn the cumulative probability at that point.

**Question 1:** The salary of 10,000 workers in Renata Ltd was approximately normally distributed with mean salary Tk. 12,000 and standard deviation salary Tk. 3,000.

- Find the number of workers receiving salary Tk 12,000 and Tk 18,000.
- Find the number of workers receiving salary between Tk. 7,000 and Tk. 10,000.
- Find the number of workers receiving salary of Tk. 16,000 or less
- Find the number of workers receiving salary of Tk. 8,500 and more
- Find the number of workers receiving salary between tk.7,000 and Tk. 14,000.
- Find the % of workers receiving salary of Tk. 20,000 or more.
- Find the % of workers receiving salary of Tk. 5,000 or less
- Find the % of workers receiving salary between Tk. 13,500 and 17,250.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.linspace(start = 1000, stop = 23000, num = 10000)
y_axis = [norm.pdf(x, loc = 12000, scale = 3000) for x in x_axis]

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Find the number of workers receiving salary b/w 12000 & 18000

mean = 12000
sd = 3000
pr = norm.cdf(x = 18000, loc = mean, scale = sd) - norm.cdf(x = 12000, loc = mean, scale = sd)
pr*10000
```

```python
# Find the number of workers receiving salary b/w 7000 & 10000

mean = 12000
sd = 3000
pr = norm.cdf(x = 10000, loc = mean, scale = sd) - norm.cdf(x = 7000, loc = mean, scale = sd)
pr*10000
```

```python
# Find the number of workers receiving salary of Tk. 16,000 or less

mean = 12000
sd = 3000
pr = norm.cdf(x = 16000, loc = mean, scale = sd)
pr*10000
```

```python
# Find the number of workers receiving salary of Tk. 8,500 and more

mean = 12000
sd = 3000
pr = 1 - norm.cdf(x = 8500, loc = mean, scale = sd)
pr*10000
```

```python
# Find the number of workers receiving salary between tk.7,000 and Tk. 14,000.

mean = 12000
sd = 3000
pr = norm.cdf(x = 14000, loc = mean, scale = sd) - norm.cdf(x = 7000, loc = mean, scale = sd)
pr*10000
```

```python
# Find the % of workers receiving salary of Tk. 20,000 or more.

mean = 12000
sd = 3000
pr = 1 - norm.cdf(x = 20000, loc = mean, scale = sd)
pr*100
```

```python
# Find the % of workers receiving salary of Tk. 5,000 or less

mean = 12000
sd = 3000
pr = norm.cdf(x = 5000, loc = mean, scale = sd)
pr*100
```

```python
# Find the % of workers receiving salary between Tk. 13,500 and 17,250.

mean = 12000
sd = 3000
pr = norm.cdf(x = 17250, loc = mean, scale = sd) - norm.cdf(x = 13500, loc = mean, scale = sd)
pr*100
```

**Question 2:** The world famous cell phone company NOKIA wishes to set a minimum hours guarantee on its new model NOK-333 a high quality cell phone. Tests reveal the mean hours are 67,900 with a standard deviation of 2,050 hours and that the distribution of hours follows the normal distribution. They want to set the minimum guaranteed hours so that no more than 4% of the cell phones will have to be replaced. What minimum guaranteed hours should NOKIA announce?

```python
mean = 67900
sd = 2050

x_axis = np.linspace(start = 57000, stop = 80000, num = 10000)
y_axis = [norm.pdf(x, loc = mean, scale = sd) for x in x_axis]

plt.plot(x_axis, y_axis)
plt.show()
```

```python
norm.cdf(x = 64299, loc = mean, scale = sd)
```

```python
norm.ppf(q = 0.04, loc = mean, scale = sd) #Inverse of CDF function
```
