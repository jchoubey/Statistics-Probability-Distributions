## The $Chi^2$ Distribution

What is a $Chi^2$ Distribution?

It is the square of a selection taken from **Standard Normal Distribution**.

<u> Intuition for the shape of $Chi^2$ distribution: </u>

- In the standard Normal distribution, 68.3% of the data is located under 1 standard deviation from the mean. (Recall that in Std Normal Dist. Mean = 0 & Std Deviation = 1). Hence, 68.3% of the data falls between -1 to 1 in standard Normal Distribution.

- When each of that 68.3% of the data is squared, it will become positive and even smaller (as it is less than 1), and scrunch around 0 to 1, with majority of it closer to 0 than 1. Therefore, we can imagine that 68.3% of the data in $Chi^2$ distribution lies b/w 0 & 1 (majority closer to 0).

- The remaining 32% of the data in Standard Normal Distribution, lies above 1. Hence, when each of that 32% of the data is squared, it will become positive and more larger and will rapidly approach $\infty$.


<u> What happens when we <b>add</b> multiple squared values from Standard Normal Distribution? </u>
- The number of additive terms in a $Chi^2$ Distribution represent the Degree of Freedom associated with the $Chi^2$ Distribution.

### Generate a $Chi^2$ Distribution

```python
import numpy as np
from scipy.stats import norm
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate a squared distribution from standard normal

normal_rv = norm.rvs(loc = 0, scale = 1, size = 10)
print(f"Normal Random Variables: {normal_rv}\n")

chi_sq_rv = np.square(normal_rv)
print(f"Chi Square Random Variables: {chi_sq_rv}")
```

```python
chi2.rvs(loc = 0, scale = 1, df = 1, size = 10)
```

### Visual Representation of a $Chi^2$ Distribution

**With 1 Degree of Freedom**

```python
norm_rv = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_sq = np.square(norm_rv)
chi_sq = chi2.rvs(loc = 0, scale = 1, df = 1, size = 100000)

fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (24, 5))
sns.histplot(norm_rv, ax = ax[0]).set(title = 'Standard Normal')
sns.histplot(norm_sq, ax = ax[1]).set(title = 'Normal Squared')
sns.histplot(chi_sq, ax = ax[2]).set(title = 'Chi-Squared')
plt.show()
```

**With 2 Degrees of Freedom**

```python
norm_rv1 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_rv2 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_sq = np.square(norm_rv1) + np.square(norm_rv2)

chi_sq = chi2.rvs(loc = 0, scale = 1, df = 2, size = 100000)

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (24, 10))
sns.histplot(norm_rv1, ax = ax[0,0]).set(title = 'Standard Normal 1')
sns.histplot(norm_rv2, ax = ax[0,1]).set(title = 'Standard Normal 2')
sns.histplot(norm_sq, ax = ax[1,0]).set(title = 'Normal Squared')
sns.histplot(chi_sq, ax = ax[1,1]).set(title = 'Chi-Squared')
plt.show()
```

**With 3 Degrees of Freedom**

```python
norm_rv1 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_rv2 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_rv3 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_sq = np.square(norm_rv1) + np.square(norm_rv2) + np.square(norm_rv3)

chi_sq = chi2.rvs(loc = 0, scale = 1, df = 3, size = 100000)

fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (27, 10))
sns.histplot(norm_rv1, ax = ax[0,0]).set(title = 'Standard Normal 1')
sns.histplot(norm_rv2, ax = ax[0,1]).set(title = 'Standard Normal 2')
sns.histplot(norm_rv3, ax = ax[0,2]).set(title = 'Standard Normal 3')
sns.histplot(norm_sq, ax = ax[1,0]).set(title = 'Normal Squared')
sns.histplot(chi_sq, ax = ax[1,1]).set(title = 'Chi-Squared')
plt.show()
```

**With 4 degrees of Freedom**

```python
norm_rv1 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_rv2 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_rv3 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_rv4 = norm.rvs(loc = 0, scale = 1, size = 100000)
norm_sq = np.square(norm_rv1) + np.square(norm_rv2) + np.square(norm_rv3) + np.square(norm_rv4)

chi_sq = chi2.rvs(loc = 0, scale = 1, df = 4, size = 100000)

fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (27, 10))
sns.histplot(norm_rv1, ax = ax[0,0]).set(title = 'Standard Normal 1')
sns.histplot(norm_rv2, ax = ax[0,1]).set(title = 'Standard Normal 2')
sns.histplot(norm_rv3, ax = ax[0,2]).set(title = 'Standard Normal 3')
sns.histplot(norm_rv4, ax = ax[1,0]).set(title = 'Standard Normal 4')
sns.histplot(norm_sq, ax = ax[1,1]).set(title = 'Normal Squared')
sns.histplot(chi_sq, ax = ax[1,2]).set(title = 'Chi-Squared')
plt.show()
```
