## Uniform Distribution

What is a Uniform Distribution?

Uniform distribution is a term used to describe a form of probability distribution where every possible outcome has an equal likelihood of happening. The probability is constant.

<u> Intuition behind the shape of Uniform Distribution. </u>

A constant line with a height = probability of outcome (as probability of each outcome is same, the line is hence constant) between the bounds of uniform distribution. Beyond the bounds, the line = 0.

### Generate a Uniform Distribution

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(1)

unif_rv = uniform.rvs(loc = 1, scale = 11, size = 10000)
print(unif_rv)
sns.histplot(unif_rv)
plt.show()
```

Observation: Every number within the specified interval has an equally likely probability of occuring.

loc = minimum value <br></br>
scale = maximum - minimum value (range)

### Visualizing Uniform Distribution

```python
x_axis = np.linspace(start = 0, stop = 12, num = 10000)
y_axis = [uniform.pdf(x, loc = 2, scale = 8) for x in x_axis]

plt.plot(x_axis, y_axis)
plt.show()
```

### Moments of a Uniform Distribution

$$Mean = \frac{a+b}{2}$$

$$Variance = \frac{(b-a)^2}{12}$$

```python
uniform.stats(loc = 2, scale = 8, moments = 'mv')
```

```python
print(f"Mean = {(2+10)/2}\n")
print(f"Var = {((10-2)**2)/12}")
```

### Example Problems:

1. The waiting time at a bus stop is uniformly distributed between 1 and 12 minute.

    - What is the probability density function?
    - What is the probability that the rider waits 8 minutes or less?
    - What is the expected waiting time?
    - What is standard deviation of waiting time?

*Solution:*

```python
import numpy as np
from scipy.stats import uniform

x_axis = np.linspace(start = 0, stop = 15, num = 10000)
y_axis = [uniform.pdf(x, loc = 1, scale = 11) for x in x_axis]

print(f"Probability of bus arriving at the bus stop b/w 1 & 12 minutes is uniformly distributed. P = {1/(12-1)}")
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Probability that the rider waits 8 min or less: Area under the curve at x = 8

uniform.cdf(x = 8, loc = 1, scale = 11)
```

```python
# Expected waiting time: Mean of the probability distribution function:

uniform.stats(loc = 1, scale = 11, moments = 'm')
```

```python
# Standard deviation of waiting time

np.sqrt(uniform.stats(loc = 1, scale = 11, moments = 'v'))
```

2. Assume the weight of a randomly chosen American passenger car is a uniformly distributed random variable ranging from 2,500 pounds to 4,500 pounds.

    - What is the mean and standard deviation of weight of a randomly chosen vehicle?
    - What is the probability that a vehicle will weigh less than 3,000 pounds?
    - More than 3,900 pounds?
    - Between 3,000 and 3,800 pounds?
    
*Solution:*

```python
x_axis = np.linspace(start = 2000, stop = 5000, num = 100000)
y_axis = [uniform.pdf(x, loc = 2500, scale = 2000) for x in x_axis]

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Mean & SD of the weight.

mean, var = uniform.stats(loc = 2500, scale = 2000, moments = 'mv')
print(f"Mean: {mean}\nSD: {var**0.5}")
```

```python
# Probability that the vehicle will weigh < 3000 pounds:

uniform.cdf(x = 3000, loc = 2500, scale = 2000)
```

```python
# Probability that the vehicle weighs > 3900 pounds:

1 - uniform.cdf(x = 3900, loc = 2500, scale = 2000)
```

```python
# Probability that the vehicle weighs b/w 3000 & 3800 pounds:

uniform.cdf(x = 3800, loc = 2500, scale = 2000) - uniform.cdf(x = 3000, loc = 2500, scale = 2000)
```
